# Combinatorial Auction Simulations

Use the helper functions in `function.py` to generate transportation graphs and bidder demand for the combinatorial auction experiments.

```python
from pprint import pprint

import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

from ACE_assignment_heuristic import run_ace_assignment
from CAV_functions import (
    generate_assignment_valuations,
    random_assignment_instance,
)
from CATS_functions import (
    generate_transport_graph,
    generate_bids,
    serialize_bids,
    draw_transport_graph,
)

# Use a consistent plotting style for notebook figures.
plt.style.use('seaborn-v0_8-colorblind')

# --- Simulation parameters -------------------------------------------------
NUM_GOODS = 10
NUM_BIDS = 5
RANDOM_SEED = 42

# --- Generate graph ---------------------------------------------------------
transport_graph = generate_transport_graph(
    num_goods=NUM_GOODS,
    seed=RANDOM_SEED,
)

print(f"Generated graph with {transport_graph.number_of_nodes()} cities and {transport_graph.number_of_edges()} goods.")

# --- Generate XOR bids ------------------------------------------------------
xor_bid_sets = generate_bids(
    transport_graph,
    num_bids=NUM_BIDS,
    seed=RANDOM_SEED,
)

pprint(serialize_bids(xor_bid_sets)[:3])
```

```python
# Visualise the generated network
_ = draw_transport_graph(
    transport_graph,
    title="Transport Network",
    show_capacities=True,
)
```

```python
# --- Generate assignment valuation instance ---------------------------------
cav = generate_assignment_valuations(
    num_bidders=6,
    num_goods=12,
    good_correlation=0.3,
    interest_model="per_item_p",
    interest_p=0.5,
    max_capacity=3,
    max_budget=20,
    rng_seed=RANDOM_SEED,
)

print(f"Assignment instance with {cav['num_bidders']} bidders and {cav['num_goods']} goods.")
print('Capacities:', cav['capacities'])
print('Budgets   :', cav['budgets'])
```

```python
# Spot-check valuations so we know the generator behaves as expected.
sample_goods = set(range(min(3, cav['num_goods'])))
print('Sample goods:', sample_goods)
print('Bidder 0 value:', cav['value_fn'](0, sample_goods))
print('Bidder 1 value:', cav['value_fn'](1, sample_goods))
```

```python
# --- Run ACE heuristic to approximate a competitive equilibrium --------------
ace_result = run_ace_assignment(
    cav,
    alpha=1.0,
    beta=1.0,
    delta=0.5,
    max_rounds=50,
    max_player_passes=2,
    epsilon_scale=0.1,
    price_floor=0.0,
    rng_seed=RANDOM_SEED,
)

print('Equilibrium reached:', ace_result['equilibrium'])
print('Rounds executed    :', ace_result['rounds'])
print('Final prices       :', ace_result['prices'])
print('Capacity adjust.   :', ace_result['capacity_adjustments'])
print('Equilibrium demands:', ace_result['demands_sorted'])
print('Budget relaxation  :', ace_result['budget_relaxation'])
print('Demand spend       :', ace_result['demand_spend'])
```

```python
# Visualise price trajectories and the final demand imbalance.
price_history = ace_result['meta']['round_history']
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

if price_history:
    price_matrix = np.vstack([entry['prices'] for entry in price_history])
    for good_idx in range(price_matrix.shape[1]):
        axes[0].plot(price_matrix[:, good_idx], label=f'Good {good_idx}')
    axes[0].set_title('Price evolution across rounds')
    axes[0].set_xlabel('Round index')
    axes[0].set_ylabel('Price')
    axes[0].legend(loc='best', fontsize=8)
else:
    axes[0].text(0.5, 0.5, 'No price updates recorded', ha='center', va='center')
    axes[0].set_axis_off()

final_counts = ace_result['meta']['demand_counts']
axes[1].bar(np.arange(len(final_counts)) - 0.2, final_counts, width=0.4, label='Demand')
axes[1].bar(np.arange(len(final_counts)) + 0.2, cav['capacities'], width=0.4, label='Capacity')
axes[1].set_title('Final demand vs. capacities')
axes[1].set_xlabel('Good index')
axes[1].set_ylabel('Count')
axes[1].legend()
plt.tight_layout()
plt.show()
```
